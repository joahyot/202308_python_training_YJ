```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# (1) 연산
# data["새로운 열의 이름"] = 연산
# 새로운 열의 이름 : xyz_mean
# 연산             : (x + y + z)/자유도 (df: degree of freedom)

diamonds["xyz_mean"] = (diamonds.x + diamonds.y + diamonds.z)/3
diamonds.head()
```

```python
# (2) 변환(Transformation)
# 숫자로 된 데이터만 적용

# 로그 변환(log), 루트 변환(root), 역수 변환(inverse)
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
# i. 로그 변환(Log Transformation)
# 새로운 열의 이름 : price_log10
# numpy.log10(수치형 데이터)

diamonds["price_log10"] = np.log10(diamonds.price)
```

```python
# ii. 루트 변환(root Transformation)
# 새로운 열의 이름 : price_root
# numpy.root(수치형 데이터)
```

```python
# diamonds["price_root"] = np.roots(diamonds.price)
# diamonds.head()
```

```python
# iii. 역수 변환(inverse Transformation)
# 새로운 열의 이름 : price_inverse
# 1/수치형 데이터
```

```python
diamonds["price_inverse"] = 1/diamonds.price
```

```python
# 로그 변환, 루트 변환, 역수 변환 : Box-Cox Transformation
```

```python
# (3) 범주(Category) 줄이기 또는 수치형 데이터를 범주형 데이터로 만들기
# cut: "Fair", "Good", "Very Good", "Premiun", "Ideal"
# 새로운 열의 이름: cut2 : "Ideal", "NonIdeal"

# data["범주형 데이터"].apply(lambda x : 조건이 맞을 때에 새로운 열이 가지는 값
#                             if 조건
#                             else 조건이 맞지 않을 때에 새로운 열이 가지는 값)
```

```python
diamonds["cut2"] = diamonds["cut"].apply(lambda x : "Ideal" if x == "Ideal" else "NonIdeal")
diamonds.head()
```

```python
# price : 수치형 데이터
# price2 : 범주형 데이터 : "Cheap", "NonCheap", price <5000

diamonds["price2"] = diamonds["price"]. apply(lambda x : "Cheap" if x < 5000 else "Noncheap")
diamonds.head()
```

```python
# cut3: "Ideal", "Premium", "ETC" (다중 if, 중첩 if)
diamonds["cut3"] = diamonds["cut"].apply(lambda x: "Ideal" if x == "Ideal" else ("Premium" if x == "Premium"   else "ETC" ))
```

```python
# 사용자 함수 만들기
# def function_name(x, y):
#     작업

# function_name(a)
```

```python
def mul_category(x): 
    if x == "Ideal":
        return "Ideal"
    elif x == "Premium" :
        return "Premium"
    else:
        return "ETC"
```

```python
mul_category("Ideal")
```

```python
mul_category("Premium")
```

```python
mul_category("Good")
```

```python
diamonds["cut4"] = diamonds["cut"].apply(mul_category)
```

```python
# (4) 구간의 정보를 갖는 새로운 열을 만들기
# 수치형 데이터를 범주형 데이터로 만들기

# 새로운 열의 이름 : price_group
# 0 이상 ~ 5000 미만 : "Very Cheap"
# 5000 이상 ~ 10000미만 : "Cheap"
# 10000 이상 ~ 15000미만 : "Expensive"
# 15000이상 ~ 20000미만 : "Very Expensive"

# pandas.cut(수치형 데이터, 
#            bins = 구간의 정보,
#            right = False or True,
#            labels = 구간의 이름)

diamonds["price_group"] = pd.cut(diamonds.price,
                                 bins = [0, 5000, 10000, 15000, 20000],
                                 right = False,
                                 labels = ["Very Cheap", "Cheap", "Expensive", "Very Expensive"])
```

```python
# numpy.arange(start = , stop = , step = )
# start부터 시작해서 stop 전까지 step 만큼 증가하는 숫자들을 만듦
# 생성한 숫자들은 bins에 활용할 수 있음

np.arange(start = 0, stop = 21000, step = 1000)
```

```python
diamonds["price_group"] = pd.cut(diamonds.price,
                                 bins = np.arange(start = 0, stop = 21000, step = 5000),
                                 right = False,
                                 labels = ["Very Cheap", "Cheap", "Expensive", "Very Expensive"])
diamonds.tail()
```

```python
# 6.7 데이터 합치기 
# 위/아래, 왼쪽/오른쪽
```

```python
# (1) 위/아래
# 각 데이터마다 열의 이름과 열의 순서도 동일해야 함
# 예제 : 2021년도 데이터, 2022년도 데이터, 2023년도 데이터

# pandas.concat(objs = [data1, data2, ...], axis = 0 )

# 참고
# 데이터는 데이터 프레임, 2차원 구조
# axis = 0 : 행
# axis = 1 : 열
```

```python
d1 = pd.DataFrame({"id": [1, 2, 3],
                  "age": [10, 20, 30],
                  "year" : [2021, 2021, 2021]})
```

```python
d2 = pd.DataFrame({"id": [4, 5, 6],
                  "age": [40, 50, 60],
                  "year" : [2022, 2022, 2022]})
```

```python
d3 = pd.DataFrame({"id": [7, 8, 9],
                  "age": [70, 80, 90],
                  "year" : [2023, 2023, 2023]})
```

```python
d4 = pd.concat(objs = [d1, d2, d3], axis = 0)
d4.reset_index() # index 다시 세팅해주는 함수
```

```python
# (2) 왼쪽/오른쪽
# 각 행은 동일한 개체의 데이터여야 함.
# pandas.concat(objs = [data1, data2, data3, ...], axis = 1)
```

```python
d5 = pd.DataFrame({"id": [1, 2, 3],
                  "name" : ["무빙", "DP2", "마스크걸"]})

d6 = pd.DataFrame({"height": [170, 163, 182],
                  "weight" : [55, 65, 100]})
```

```python
d7 = pd.concat(objs=[d5, d6], axis = 1)
d7
```

```python
# (3) 왼쪽/오른쪽
# 2개의 데이터만 합칠 수 있음
# 왼쪽 데이터의 primary key가 있고
# 오른쪽 데이터의 primary key가 있을 때에 가능함

# primary key: 기본키, 주요키 
# 왼쪽 , 오른쪽의 primary key 이름이 동일해야 함

# Merge or Join 
```

```python
d8 = pd.DataFrame({"id": [1, 2, 4, 5],
                  "age": [20, 40, 50, 60]})
```

```python
d9 = pd.DataFrame({"id": [1, 3,  4, 6],
                  "bt": ["A", "B", "AB", "O"]})
```

```python
# i. Inner Join 
# 2개의 Primary key가 동일한 데이터만 합침
# pandas.merge(data1, data2, on = "primary key", how = "inner")
```

```python
d10 = pd.merge(d8, d9, on = "id", how = "inner")
d10
```

```python
# ii. outer join : full join 
# 왼쪽, 오른쪽의 Primary key의 합집합
# pandas.merger(data1, data2, on = "primary key", how = "outer")
```

```python
d11 = pd.merge(d8, d9, on = "id", how = "outer")
d11.sort_values(by = "id", ascending = True)
```

```python
# iii. outer join : left join 
# 왼쪽에 있는 Primary key의 합집합
# pandas.merger(data1, data2, on = "primary key", how = "left")
```

```python
d12 = pd.merge(d8, d9, on = "id", how = "left")
d12
```

```python
# iv. outer join : right join 
# 오른쪽에 있는 Primary key의 합집합
# pandas.merger(data1, data2, on = "primary key", how = "right")
```

```python
d13 = d12 = pd.merge(d8, d9, on = "id", how = "right")
d13
```
